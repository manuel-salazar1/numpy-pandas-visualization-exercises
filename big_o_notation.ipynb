```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
# X-Values
n = np.arange(1,20)
```

```python
#O(1)
y1 = 0 * n + 1
#O(log n)
y2 = np.log(n)
#O(n)
y3 = n
# O(n log n)
y4 = np.log(n) * n
# O(n^2)
y5 = n**2
# O(2^n)
y6 = 2**n
# O(n!)
y7 = [math.factorial(i) for i in n]
# O(n^n)
y8 = n**n
```

```python
plt.figure(figsize = (16, 8))

plt.plot(n, y1, label='$\mathcal{O}(1)$')
plt.plot(n, y2, label='$\mathcal{O}(logn)$')
plt.plot(n, y3, label='$\mathcal{O}(n)$')
plt.plot(n, y4, label='$\mathcal{O}(nlogn)$')
plt.plot(n, y5, label='$\mathcal{O}(n^{2})$')
plt.plot(n, y6, label='$\mathcal{O}(2^{n})$')
plt.plot(n, y7, label='$\mathcal{O}(n!)$')
plt.plot(n, y8, label='$\mathcal{O}(n^{n})$')

plt.xlim(1, 11)
plt.ylim(0, 80)

plt.xlabel('Elements')
plt.ylabel('Operations')

plt.title('Big O Notation')

plt.legend()
plt.show()
```
